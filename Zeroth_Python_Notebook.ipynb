```python
city = 'Los Angeles'

data = pd.read_csv('Los Angeles-Office-1-1.csv', parse_dates=False)
df = pd.DataFrame({'datetime': pd.date_range('2019-01-01', '2020-01-01', freq='.1H', closed='left')})
data['datetime'] = (df['datetime'])
data = data.set_index(['datetime'])

data.head()
```

```python
#defining humidity ratio for psychrometric plotting

data['Tair_Indoors'] = data['Tair_Indoors                            ']
data['Tair_Outdoors'] = data['Tair_Outdoors                           ']
data['Relative_Humidity'] = data['Indoor_RH                               ']
data['Relative_Humidity_Outdoors'] = data['Outdoor_RH                              ']
data['HR']=data['Relative_Humidity']*10*0.62198*np.exp(77.345+0.0057*(data['Tair_Indoors']+273.15)-7235/(data['Tair_Indoors']+273.15))/(101325*np.power((data['Tair_Indoors']+273.15),8.2)-np.exp(77.345+0.0057*(data['Tair_Indoors']+273.15)-7235/(data['Tair_Indoors']+273.15)))
data['HR_outside']=data['Relative_Humidity_Outdoors']*10*0.62198*np.exp(77.345+0.0057*(data['Tair_Outdoors']+273.15)-7235/(data['Tair_Outdoors']+273.15))/(101325*np.power((data['Tair_Outdoors']+273.15),8.2)-np.exp(77.345+0.0057*(data['Tair_Outdoors']+273.15)-7235/(data['Tair_Outdoors']+273.15)))

data['Month'] = pd.DatetimeIndex(data.index).month
data['Month'] = data['Month'].apply(lambda x: calendar.month_abbr[x])

data['Day'] = pd.DatetimeIndex(data.index).day_name()
data['Hour'] = pd.DatetimeIndex(data.index).hour
 
data = data.resample('60T').mean()
```

```python
ax = plt.subplot(111)
data.plot.scatter(x='Tair_Indoors', y='HR', c='r', alpha = 0.1, figsize=(11,7), label = 'Indoor Air', ax=ax)
data.plot.scatter(x='Tair_Outdoors', y='HR_outside', c='k', alpha = 0.1, figsize=(11,7), label = 'Outdoor Air', ax=ax)
#data.plot.scatter(x='HouDB', y='HR_Hou', c='r', alpha = 0.1, figsize=(11,7), label = 'Outdoor Air', ax=ax)
#data.plot.scatter(x='AtlDB', y='HR_Atl', c='r', alpha = 0.1, figsize=(11,7), label = 'Outdoor Air', ax=ax)
#data.plot.scatter(x='Bal DB', y='HR_Bal', c='r', alpha = 0.1, figsize=(11,7), label = 'Outdoor Air', ax=ax)
#data.plot.scatter(x='Chi DB', y='HR_Chi', c='r', alpha = 0.1, figsize=(11,7), label = 'Outdoor Air', ax=ax)
#data.plot.scatter(x='MinnDB', y='HR_Minn', c='r', alpha = 0.1, figsize=(11,7), label = 'Outdoor Air', ax=ax)
#df100.loc[(df100['Load'] !=0)].plot.scatter(x='Tdb,i', y='g,i', c='k', alpha = 0.1, label = 'Room Air', ax=ax)
#df100.loc[(df100['Load'] !=0)].plot.scatter(x='Tdb,o', y='g,o', c='g', alpha = 0.1, label = 'Supply Air', ax=ax)

#brute force psychrometric plotting
ax.plot(temperature, saturation, 'k-', linewidth = 1)
ax.plot(temperature, saturation*.9, 'k-', linewidth = 1)
ax.plot(temperature, saturation*.8, 'k-', linewidth = 1)
ax.plot(temperature, saturation*.7, 'k-', linewidth = 1)
ax.plot(temperature, saturation*.6, 'k-', linewidth = 1)
ax.plot(temperature, saturation*.5, 'k-', linewidth = 1)
ax.plot(temperature, saturation*.4, 'k-', linewidth = 1)
ax.plot(temperature, saturation*.3, 'k-', linewidth = 1)
ax.plot(temperature, saturation*.2, 'k-', linewidth = 1)
ax.plot(temperature, saturation*.1, 'k-', linewidth = 1)
#ax.set_xlim(10,45)
ax.set_ylim(0,25)
ax.set_xlabel('Temperature ($\degree$C)', size=textsize)
ax.set_ylabel('Humidity Ratio (g/kg)', size=textsize)
ax.legend(loc=2, fontsize=textsize)
plt.tick_params(axis='both', labelsize=textsize)

plt.savefig('Psychro.png', format='png')
```

```python
df = data
df['date'] = df.index.normalize()
df['time'] = df.index.time
df['HR_difference'] = df['HR']-df['HR_outside']
df_pivot = df.pivot(index='date', columns='time', values='Active_Vent_Flow_Kg/hr                  ')
df_pivot_solar = df.pivot(index='date', columns='time', values='Humidity_From_Gains                     ')
df_pivot_HR = df.pivot(index='date', columns='time', values='HR_difference')
df_pivot_cold_delivered = df.pivot(index='date', columns='time', values='ColdDelivBuild                          ')
df_pivot_cold_generated = df.pivot(index='date', columns='time', values='ColdCreated                             ')

df_pivot_heat_delivered = df.pivot(index='date', columns='time', values='HeatDelivBuild                          ')
df_pivot_heat_generated = df.pivot(index='date', columns='time', values='Heat_Created                            ')

fig, ax = plt.subplots(1, 1, figsize=(10, 6))
df_pivot.T.plot(ax=ax, alpha=.1, color = 'k', legend = None)
df_pivot.mean().T.plot(ax=ax, alpha=1, color = 'r', legend = None)


plt.ylabel('ventilation Air flow (m$^3$/hr)', size = textsize)
plt.xlabel('Time of Day', size = textsize)
plt.tick_params(axis='both', labelsize=textsize)

plt.savefig("dailyFan.png", format='png')
```
